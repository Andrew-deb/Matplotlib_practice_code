```python
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
#To create a pie chart we use the pie method.
slices= [60, 40]
plt.pie(slices)
plt.title('My awesome pie chart')
plt.tight_layout()
plt.show()
```

```python
#To make our pie chart more descriptive, we need to add some labels
slices= [60, 40]
labels= ['Sixty', 'Fourty'] #Note that the pie cart doesn't always sum up to 100% rather it takes the proportion of the total number of elements available
plt.pie(slices, labels= labels, wedgeprops= {'edgecolor': 'black'}) #The wedgeprops argument helps to brighten the edges
plt.title('My awesome pie chart')
plt.tight_layout()
plt.show()
```

```python
#We could also change the color of the slices, by creating a list of colors according to the respective slices we want them to be assigned to.
#Colors:
Blue= '#008fd5'
Red= '#fc4f30'
Yellow= '#e5ae37'
Green= '#6d904f'

slices= [60, 40, 30, 20]
labels= ['Sixty', 'Fourty', 'Extra1', 'Extra2']
color= [Blue, Red, Yellow, Green] #Note you could either pass the color codes directly to the list or save them in a variable then pass them into the list
#color= ['blue', 'red', 'yellow', 'green'] #These are default color specifications that you could use
plt.pie(slices, colors= color, labels= labels, wedgeprops= {'edgecolor': 'black'})
plt.title('My awesome pie chart')
plt.tight_layout()
plt.show()
```

```python
languages= ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
popularity= [59219, 55466, 47544, 36443, 35917]
```

```python
#Note it's best us pie chart when yo have 5 or less values to plot.
plt.pie(popularity, labels= languages, wedgeprops= {'edgecolor': 'black'})
plt.title('My awesome pie chart')
plt.tight_layout()
plt.show()
```

```python
#Assume we want to emphasize more on a slice of the pie chart, we could the use the explode argument
explode= [0, 0, 0, 0.1, 0]
plt.pie(popularity, labels= languages, wedgeprops= {'edgecolor': 'black'}, explode= explode)
plt.title('My awesome pie chart')
plt.tight_layout()
plt.show()
```

```python
#We could also pass the shadow argument to give it a 3d-look
plt.pie(popularity, labels= languages, wedgeprops= {'edgecolor': 'black'}, explode= explode, shadow= True)
plt.title('My awesome pie chart')
plt.tight_layout()
plt.show()
```

```python
#To rotate the position of each slice on the pie chart we use the startangle argument
plt.pie(popularity, labels= languages, wedgeprops= {'edgecolor': 'black'}, explode= explode, shadow= True, startangle= 90)
plt.title('My awesome pie chart')
plt.tight_layout()
plt.show()
```

```python
#To display the percentage each slice takes in the whole pie chart, we use the autopct argument
plt.pie(popularity, labels= languages, wedgeprops= {'edgecolor': 'black'}, 
        explode= explode, autopct='%1.1f%%',
        shadow= True, startangle= 90)
plt.title('My awesome pie chart')
plt.tight_layout()
plt.show()
```
