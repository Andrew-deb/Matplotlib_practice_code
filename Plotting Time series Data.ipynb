```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
from datetime import datetime, timedelta
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
dates= [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]
```

```python
y= [0, 1, 3, 4, 6, 5, 7]
```

```python
#We use the .plot_date() method variable to plot datesets with dates.
plt.plot_date(dates, y)
plt.tight_layout()
plt.show()
```

```python
#We pass the linestyle argument to connect the points together
plt.plot_date(dates, y, linestyle= 'solid')
plt.tight_layout()
plt.show()
```

```python
#We could also turn of the markers which indicates the date points in our plot
plt.plot_date(dates, y, linestyle= 'solid', marker= None) #Here we set the marker argument to None to remove the markers
plt.tight_layout()
plt.show()
```

```python
#Now, we could also change the alignment of the dates in the x-axis, ie. rotation
plt.plot_date(dates, y, linestyle= 'solid')
plt.gcf().autofmt_xdate() #.gcf() meaning get current figure and the .autofm_xdate() method id helps to align each figure in the x-axis
plt.tight_layout()
plt.show()
```

```python
#We could also change the format of our date using the mpl_dates.DateFormatter('%b', %d %Y)
#Where %b serves as a place holder for the month, so as the %d for days and the %Y for years
plt.plot_date(dates, y, linestyle= 'solid')
plt.gcf().autofmt_xdate()

date_format= mpl_dates.DateFormatter('%b, %d %Y')
plt.gca().xaxis.set_major_formatter(date_format) #gca means get current axis.

plt.tight_layout()
plt.show()
```

```python
data= pd.read_csv('Data\data.csv')
price_date= data['Date']
price_close= data['Close']
```

```python
#Note the price date  show the price of bitcoin at that particular day and the price close show theprice of bitcoin at the end of that day.
plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()
plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.tight_layout()
plt.show()
```
