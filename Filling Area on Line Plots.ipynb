```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
dev_data= pd.read_csv('Data/filling between data.csv')
```

```python
ages= dev_data['Age']
dev_salaries= dev_data['All_Devs']
pyDev_salaries= dev_data['Python']
jsDev_salaries= dev_data['JavaScript']
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle= '--', label='All Devs')
plt.plot(ages, pyDev_salaries, label= 'Python')

#overall_median= 57287
plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
#Adding fills to our plot using the .fill_between() method.
plt.plot(ages, dev_salaries, color='#444444', linestyle= '--', label='All Devs')
plt.plot(ages, pyDev_salaries, label= 'Python')

#overall_median= 57287

plt.fill_between(ages, pyDev_salaries, alpha= 0.25) #The alpha arg gives the fill color a brignt color.
plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
#Note that in this case our fill_between arg is defaulting to 0, therefore filling from our python plot up to 0 on the graph.
```

```python
#Now we could fill between the median salaries of all developes instead of using zero, by passing the median salaries of all developres into the fill_between method as a y-value argument.
plt.plot(ages, dev_salaries, color='#444444', linestyle= '--', label='All Devs')
plt.plot(ages, pyDev_salaries, label= 'Python')

overall_median= 57287

plt.fill_between(ages, pyDev_salaries, overall_median, alpha= 0.25) #The alpha arg gives the fill color a brignt color.
plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
#Again, we could change the color both at the above and below the median salary indication, by passing a where argument into the fill_between method.
#Now we could fill between the median salaries of all developers instead of using zero, by passing the median salaries of all developers into the fill_between method as a y-value argument.
plt.plot(ages, dev_salaries, color='#444444', linestyle= '--', label='All Devs')
plt.plot(ages, pyDev_salaries, label= 'Python')

overall_median= 57287

plt.fill_between(ages, pyDev_salaries, overall_median, alpha= 0.25, where=(pyDev_salaries > overall_median), interpolate= True) #Here within the fill_between method we pass the where arg and specified with conditionals where the color should take effect.
#Also the interpolate argument makes sure that certain x intersections don't get clipped and all of the regions are been filled correctly.
plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
#To make the fill color between the two segment of the plot we have to create two fill_between methods and pass different colors to represent different regions of the graph
plt.plot(ages, dev_salaries, color='#444444', linestyle= '--', label='All Devs')
plt.plot(ages, pyDev_salaries, label= 'Python')

overall_median= 57287

plt.fill_between(ages, pyDev_salaries, overall_median, alpha= 0.25, where=(pyDev_salaries > overall_median), interpolate= True)
plt.fill_between(ages, pyDev_salaries, overall_median, alpha= 0.25, where=(pyDev_salaries <= overall_median), interpolate= True)
plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
#We could also customize the colors at any region, by passing in a color argument.
plt.plot(ages, dev_salaries, color='#444444', linestyle= '--', label='All Devs')
plt.plot(ages, pyDev_salaries, label= 'Python')

overall_median= 57287

plt.fill_between(ages, pyDev_salaries, overall_median, alpha= 0.25, where=(pyDev_salaries > overall_median), interpolate= True)
plt.fill_between(ages, pyDev_salaries, overall_median, alpha= 0.25, where=(pyDev_salaries <= overall_median), interpolate= True, color='red')
plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
#We could also label each section of our graph to give the viewers more understanding of what the graph is emphazing on, by passing the label argument.
plt.plot(ages, dev_salaries, color='#444444', linestyle= '--', label='All Devs')
plt.plot(ages, pyDev_salaries, label= 'Python')

overall_median= 57287

plt.fill_between(ages, pyDev_salaries, overall_median,
                 alpha= 0.25, where=(pyDev_salaries > overall_median), 
                 interpolate= True, label= 'Above Average')

plt.fill_between(ages, pyDev_salaries, overall_median, 
                 alpha= 0.25, where=(pyDev_salaries <= overall_median), 
                 interpolate= True, label='Below Average')
plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
import random
from itertools import count
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
#We could create a plot of data that continuous changes frequently in real time, which would be great to represent data that changes frequently that we want to monitor.
x_vals= []
y_vals= []
plt.plot(x_vals, y_vals)

index= count()

def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0, 5))

    plt.plot(x_vals, y_vals)
    
ani= FuncAnimation(plt.gcf(), animate, interval= 1000, save_count=10 ) 
#The FuncAnimation() method takes the plt.gcf() method to get the current figure we are plotting on, the animate arg from the matplotlib.animation library which adjust the graph as the data changes and the interval arg to specify the time interval for each adjustment.
 
plt.tight_layout()
plt.show()
```
