```python
from itertools import count
from matplotlib.animation import FuncAnimation
import matplotlib.pyplot as plt
import random

x_vals = []
y_vals = []

index = count()

# Initialize an empty line object
line, = plt.plot([], [], marker='o', linestyle='None')  # Use markers to show individual data points

def animate(i):
    x = next(index)
    y = random.randint(0, 5)
    
    # Append new data point
    x_vals.append(x)
    y_vals.append(y)
    
    # Update the data of the line object
    line.set_data(x_vals, y_vals)
    
    # Adjust plot limits dynamically
    plt.xlim(min(x_vals), max(x_vals))
    plt.ylim(min(y_vals), max(y_vals))

ani = FuncAnimation(plt.gcf(), animate, interval=1000, save_count= 10)

plt.tight_layout()
plt.show()
```
