```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('seaborn')
```

```python
x= [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y= [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

```python
#colors= [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
#sizes= [209, 486, 381, 255, 191, 315, 185, 228, 174, 538, 239, 399, 153, 273, 293, 436, 501, 397, 539]
```

```python
#To create a scatter plot we use the scatter() method
plt.scatter(x, y, s=100, ) #The s argument is used to resize the size of all the points on the scatter plot
plt.tight_layout()
plt.show()
```

```python
#We could also adjust the marker and color type by passing in a color and marker argument.
plt.scatter(x, y, s=100, color= 'green', marker='x')
plt.tight_layout()
plt.show()
```

```python
#We could also brighten the edges of each point in the scatter plot by pasing in the edgecolor and linewidth arg to brighten and sharpen the edge of each points respectively
plt.scatter(x, y, s=100, edgecolor= 'black', linewidth=1, alpha= 0.75) #The alpha argument helps in making our points more transparent.
plt.tight_layout()
plt.show() 
```

```python
#We could pass a color list which would have the same length with the list of values contained at both x and y and define the intensity of the color at every point at which x and y intersects

colors= [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
plt.scatter(x, y, s=100, c=colors, edgecolor= 'black', linewidth=1, alpha= 0.75)
plt.tight_layout()
plt.show() 
```

```python
#We could change or adjust the color of used at each point of intersection, passing the cmap argument
colors= [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
plt.scatter(x, y, s=100, c=colors, cmap='Greens', edgecolor= 'black', linewidth=1, alpha= 0.75)


plt.tight_layout()
plt.show()
```

```python
#We could also add labels for our color maps to help the user identify different color intensity represented by the various points in our scatter plot
#To do this we use the .colorbar() method
colors= [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
plt.scatter(x, y, s=100, c=colors, cmap='Greens', edgecolor= 'black', linewidth=1, alpha= 0.75)

cbar= plt.colorbar()
cbar.set_label('Satisfation level') #The .set_label() method is a function of the .colorbar method with helps to describe what the color bar is reflecting.
plt.tight_layout()
plt.show()
```

```python
#We could also pass in a list of various sizes to adjust the size of the varios point is the scatter plot pending on what we are the project we are working on
colors= [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
sizes= [209, 486, 381, 255, 191, 315, 185, 228, 174, 538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
plt.scatter(x, y, s=sizes, c=colors, cmap='Greens', edgecolor= 'black', linewidth=1, alpha= 0.75)

cbar= plt.colorbar()
cbar.set_label('Satisfation level')
plt.tight_layout()
plt.show()
```

```python
#Now from the above scatter plot the we assigned larger point sizes to points that indicate higher satisfaction level and vise-versa.
```

## Working with Larger data set

```python
data= pd.read_csv('Data/2019-05-31-data.csv')
view_count= data['view_count']
likes= data['likes']
ratio= data['ratio']
```

```python
plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)

#cbar= plt.colorbar()
#cbar.set_label('')
plt.title('Trending YouTube Videos')
plt.xlabel('View counts')
plt.ylabel('Total Likes')
plt.tight_layout()
plt.show()
```

```python
#From this data plotted on our scatter plot, we can see that most of our points are clustered and if we seek to get a clearer view of the clustered points we could represent them in a logarithm scale
plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)

#cbar= plt.colorbar()
#cbar.set_label('')
plt.xscale('log')
plt.yscale('log')
plt.title('Trending YouTube Videos')
plt.xlabel('View counts')
plt.ylabel('Total Likes')
plt.tight_layout()
plt.show()
```

```python
#We could also pass the ratio of the likes to dislikes for YouTube videos to the c argument
plt.scatter(view_count, likes, c=ratio, cmap='summer', edgecolor='black', linewidth=1, alpha=0.75)

cbar= plt.colorbar()
cbar.set_label('Like to dislike ratio')
plt.xscale('log')
plt.yscale('log')
plt.title('Trending YouTube Videos')
plt.xlabel('View counts')
plt.ylabel('Total Likes')
plt.tight_layout()
plt.show()
```

```python
#Note that from the scatter plot if we have a data point with low color intensity and they are high up the total likes scale that implies that the videos with that data point has as much likes and dislikes for that particular video
```
