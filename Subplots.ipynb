```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
data=pd.read_csv('Data/filling between data.csv')
ages= data['Age']
dev_salaries= data['All_Devs']
py_salaries= data['Python']
js_salaries= data['JavaScript']
```

```python
plt.plot(ages, py_salaries, label= 'Python')
plt.plot(ages, js_salaries, label= 'JavaScript')

plt.plot(ages, dev_salaries, color= '#444444', linestyle= '--', label='All Devs')

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.tight_layout()
plt.show()
```

```python
#Figures are the graphs that holds our plot, they are like containers which our plot is contained or plotted in 
#WHILE our axis are the actual plot. A figure can have multiple plots and we can also have multiple figures.
```

```python
#We could do this(ie create multiple figures with different plots at the same time) either by using the plt.gcf() and plt.gca() method to switch between different figures
#OR by using the OOPs approach, which is the subplot method to instantiate a  figure and an axis.

fig, ax= plt.subplots() 
#From the above code our ax represents our axis. It is set to a single axis or plot, because by default the subplot method create a figure and then specify a certain number of row and column of axis which is 1-by-1 which is simply one axis.
#The figure specifies the number of windows our figures would be plotted on.
#Now we can replace our plt object with our ax object to our .plot method.

ax.plot(ages, py_salaries, label= 'Python')
ax.plot(ages, js_salaries, label= 'JavaScript')

ax.plot(ages, dev_salaries, color= '#444444', linestyle= '--', label='All Devs')

ax.legend()
ax.set_title('Median Salary (USD) by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Medai Salary (USD)')

plt.tight_layout()
plt.show()
```

```python
#Explaining a bit on the fig and ax objects
#Currently, in the above subplot we are using the default plot of 1 column and 1 row which gives us exactly one graph with the specificed plot.
#Now we could adjust this by passing an nrows and ncols arg to adjust the rows and the columns within our axis method respectively.
#For example:
fig, ax= plt.subplots(nrows= 2, ncols= 1) #Which gives 2 figures with two axis.
print(ax)
```

```python
fig, (ax1, ax2)= plt.subplots(nrows= 2, ncols= 1) #We could also unpack the list, to specify which axis we want to plot on.
print(ax1)
print(ax2)
```

```python
#Plotting this on our data.
fig, (ax1, ax2)= plt.subplots(nrows= 2, ncols= 1) 

ax2.plot(ages, py_salaries, label= 'Python')
ax2.plot(ages, js_salaries, label= 'JavaScript')

ax1.plot(ages, dev_salaries, color= '#444444', linestyle= '--', label='All Devs')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
#ax2.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Medai Salary (USD)')

plt.tight_layout()
plt.show()
```

```python
#We could make the x-axis and y-axis share same axis numbering by passing the sharex and sharey arg respectively, pending on the data we are working with.
#Plotting this on our data.
fig, (ax1, ax2)= plt.subplots(nrows= 2, ncols= 1) 

ax2.plot(ages, py_salaries, label= 'Python')
ax2.plot(ages, js_salaries, label= 'JavaScript')

ax1.plot(ages, dev_salaries, color= '#444444', linestyle= '--', label='All Devs')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
#ax2.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Medai Salary (USD)')

plt.tight_layout()
plt.show()
```

```python
#Assume we want our plot to reflect on two different figures(Windows), we simply make adjustment to the figure object
fig1, ax1= plt.subplots() 
fig2, ax2= plt.subplots() 

ax2.plot(ages, py_salaries, label= 'Python')
ax2.plot(ages, js_salaries, label= 'JavaScript')

ax1.plot(ages, dev_salaries, color= '#444444', linestyle= '--', label='All Devs')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Medai Salary (USD)')

plt.tight_layout()
plt.show()

fig1.savefig('fig1.png')
fig2.savefig('fig2.png')
```
